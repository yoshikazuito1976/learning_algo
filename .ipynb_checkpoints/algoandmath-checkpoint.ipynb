```python
import numpy as np
import matplotlib.pyplot as plt

# x を -2 から 10 までの範囲で生成
x = np.linspace(0,10 )

# y = min(x, 5) を要素ごとに適用
y = np.minimum(x, 5)

# グラフを描画
plt.plot(x, y, label='y = min(x, 5)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = min(x, 5)')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

# x軸の値（対数の定義域：x > 0）
x = np.linspace(0.1, 10, 400)

# 各種対数関数
y_ln = np.log(x)          # 自然対数 ln(x)
y_log10 = np.log10(x)     # 常用対数 log₁₀(x)
y_log2 = np.log2(x)       # 底2の対数 log₂(x)

# グラフの描画
plt.figure(figsize=(8, 6))
plt.plot(x, y_ln, label="ln(x)", linewidth=2)
plt.plot(x, y_log10, label="log₁₀(x)", linestyle='--')
plt.plot(x, y_log2, label="log₂(x)", linestyle=':')

plt.title("対数関数のグラフ")
plt.xlabel("x")
```

```python
def subset_sum(N, S, A):
    dp = [[False] * (S + 1) for _ in range(N + 1)]
    dp[0][0] = True  # 0枚で0を作るのは常に可能
    print(dp)
    print(A)

    for i in range(N):
        for j in range(S + 1):
            if dp[i][j]:
                dp[i + 1][j] = True  # カードiを使わない
                print(dp)
                if j + A[i] <= S:
                    dp[i + 1][j + A[i]] = True  # カードiを使う
                    print(dp)

    return "Yes" if dp[N][S] else "No"

# 入力例
N = 3
S = 11
A = [2, 5, 9]

print(subset_sum(N, S, A))  # 出力: Yes
```

```python
def bit_full_search(N, S, A):
    for bit in range(1 << N):  # 全ての組み合わせ 0〜2^N-1
        total = 0
        for i in range(N):
            if bit & (1 << i):  # i番目のカードを使う
                total += A[i]
        if total == S:
            return "Yes"
    return "No"

print(bit_full_search(3, 11, [2, 5, 9]))  # Yes
```

## bit列でon/off判定ができる（全検索部分和）
01の組み合わせで選択、非選択の判定ができる。Linuxのパーミッション設定と同じ  
全選択になるので、20ビットが限界。PCが唸る  
20ビットを超える組み合わせの場合は動的計画法をマスターしてそれを用いるのが正解  
組み合わせ部分和の計算の初歩。  
組み合わせをbit列で表現できることに感動。  
点が線になった。  

```python
"""
bit列による選択を一つ一つ確認する全検索
部分和の問題
リストaの数字の組み合わせでsの数字を表現できるかどうかを判定する問題
全検索を用いているので、n=20が限界
下のコードでは15になる例があるので、表現できると判定される
5ビットなので32通りの組み合わせがあり、全部を探索し、15になる例を探している。
20組目でOK判定がでていた。OKの時点で計算をやめれば少しは節約できるかも・・・。
"""

n=5
s=15
a=[3, 7, 1, 8, 5]
print(f'a:{a}')
for bit in range(1<<n):
    print(f'{bit+1}回目')
    print(f'bit:{bit:0{n}b}')
    total=0
    print(f'a:{a}')
    for i in range(n):
        
        if bit & (1<<i):
            #print(f'i:{i:0{n}b}')
            print(f'a[{i}]:{a[i]}')
            total += a[i]
            print(f'total:{total}')
    if total==s:
        print('OK')
    else:
        print('NG')
    print()
```
